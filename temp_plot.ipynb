```python
import matplotlib.pyplot as plt
import numpy as np

#1.  Define the data for the graphs

# For the increase and decrease phase
rate = 8/3  # temperature change rate in °C/min
time_increase_1 = int(800 / rate)  # time taken to reach 800°C
time_increase_2 = int(865 / rate)  # time taken to reach 865°C

# Plot 1 Sintering Time Frame (First Run)
x1 = [0, time_increase_1, time_increase_1 + 30*60, 2*time_increase_1 + 30*60]
y1 = [0, 800, 800, 0]

# Plot 2 Sintering Time Frame (Second Run)
x2 = [0, time_increase_2, time_increase_2 + 50*60, 2*time_increase_2 + 50*60]
y2 = [0, 865, 865, 0]



#2.  Plotting the graphs

# Plot 1
plt.figure(figsize=(10, 5))
plt.plot(x1, y1, 'r--', label='T/t = 8/3')
plt.xlabel('Time(min)')
plt.ylabel('Temperature(°Celsius)')
plt.title('Plot 1: Sintering Time Frame (First Run)')
plt.legend()
plt.grid(True)
plt.show()

# Plot 2
plt.figure(figsize=(10, 5))
plt.plot(x2, y2, 'b--', label='T/t = 8/3')
plt.xlabel('Time(min)')
plt.ylabel('Temperature(°Celsius)')
plt.title('Plot 2: Sintering Time Frame (Second Run)')
plt.legend()
plt.grid(True)
plt.show()

# Output the data for verification
print("Data for plot1:\nTime(min):", x1, "\nTemperature(°Celsius):", y1)
print("\nData for plot1:\nTime(min):", x2, "\nTemperature(°Celsius):", y2)
```

```python
# 3. Plotting the graphs on the same figure
plt.figure(figsize=(12, 6))
plt.plot(x1, y1, 'r:', label='First Run: T/t = 8/3', linewidth=2)
plt.plot(x2, y2, 'b-.', label='Second Run: T/t = 8/3', linewidth=2)
plt.xlabel('Time(min)')
plt.ylabel('Temperature(°Celsius)')
plt.title('Sintering Time Frames')
plt.legend()
plt.grid(True)
plt.show()
```
